### Question 1

**User info
What are the counts of each user type?
What are the counts of each gender (only available for NYC and Chicago)?
What are the earliest, most recent, most common year of birth (only available for NYC and Chicago)?
**

```r
# Your solution code goes here
##create 'Gender' and 'Birth.Year' columns for Washington

wash$Gender <- c("")
wash$Birth.Year <- c("")

##create 'city' column for all tables
wash$city <- c('Washington')
ny$city <- c('New York')
chi$city <- c('Chicago')

head(wash)
head(ny)
head(chi)
```

```r
## Combine all tables into one
bikeshare_cities <- rbind(ny,wash,chi)
bikeshare_cities
```

## Solution to Question 1

```r
##Visualize on a Bar Chart
table(bikeshare_cities$User.Type) ##User Type Count
ggplot(subset(bikeshare_cities, User.Type %in% c('Customer', 'Subscriber')), aes(x = User.Type, fill = User.Type)) + 
geom_bar(stat = 'count') + 
scale_y_continuous(breaks = seq(0, 125000, 25000)) + 
theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Number of Users') +
    xlab('User Type') +
    ylab('Count')
```

```r
table(bikeshare_cities$Gender) ##Gender Count
ggplot(subset(bikeshare_cities, Gender%in%c('Male', 'Female') ), aes(x = Gender, fill = Gender)) + 
geom_bar(stat = 'count') + 
scale_y_continuous(breaks = seq(0, 50000, 5000)) + 
theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Gender Distribution of Users in New York and Chicago') +
    xlab('Gender') +
    ylab('Count')
```

```r
#What are the earliest, most recent, most common year of birth (only available for NYC and Chicago)?
summary(ny$Birth.Year)
summary(chi$Birth.Year)
ggplot(subset(bikeshare_cities, na.omit = TRUE)) +
aes(x = Birth.Year, stat = 'count') + 
geom_bar() + 
theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Number of Users By Birth Year') +
    xlab('Birth Year') +
    ylab('Count')
```

```r
max(bikeshare_cities$Birth.Year[!is.na(bikeshare_cities$Birth.Year)])
```

**Summary of your question 1 results goes here.**
From the summary and chart, the number of Non-defined user, customers and subscribers for the three cities are 121, 30754 and 121576 respectively. Thus, we can deduce that the user type 'Subscriber' has the highest patronage of the bike share services in the three (3) cities which is quite great as they are repeat customers.

From the summary and chart, the number of females and males for the New York and Chicago are 13882 and 42360 respectively. Thus, we can deduce that the gender 'male' has the highest patronage of the bike share services in the two (2) cities.

From the summaries, the earliest birth year for New York and Chicago is 1885 and 1899 respectively while most common year is 2002.

### Question 2

**#3 Trip duration

What is the total travel time for users in different cities?
What is the average travel time for users in different cities?
**

```r
# Your solution code goes here
##Construct a Boxplot for Trip Duration
qplot(x = User.Type, y = Trip.Duration,
     data = subset(bikeshare_cities, User.Type!=''),
     geom = 'boxplot',
     ylab = 'Trip Duration',
     main = 'Trip Duration by City and User Type') + 
     coord_cartesian(ylim = c(0,5000)) + 
     facet_wrap( ~ city, scales = 'free')
```

```r
##Generate a chart for the average travel time for users in different cities?
by(bikeshare_cities$Trip.Duration, bikeshare_cities$city, mean, na.rm = TRUE)
ggplot(bikeshare_cities) + 
  geom_bar(aes(city, Trip.Duration, fill = as.factor(city)), 
        position = "dodge", stat = "summary", fun.y = "mean") + 
        ggtitle('Average Trip Duration for Users in the three cities') + 
        xlab('Cities') + 
        ylab('Average Trip Duration')
```

**Summary of your question 2 results goes here.**
From the barcharts, we can deduce that the city of Washington has the highest Total Travel Time & Average Travel Time for users of 109,883,545 seconds & 1,233.939 seconds respectively.

### Question 3

**2 Popular stations and trip
What is the most common start station?
What is the most common end station?
What is the most common trip from start to end (i.e., most frequent combination of start station and end station)?**

```r
# Your solution code goes here
##Generate a barchart for Start Station
top5start_station <- tail(names(sort(table(bikeshare_cities$Start.Station))), 5)
common_start_station <- subset(bikeshare_cities, Start.Station %in% top5start_station)
common_start_station$Start.Station <- factor(common_start_station$Start.Station, levels = rev(top5start_station))

ggplot(common_start_station, aes(x = Start.Station, stat = 'count', fill = Start.Station)) + 
geom_bar() + 
theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Most Common Start Stations') +
    xlab('Start station') +
    ylab('Count')
head(summary(bikeshare_cities$Start.Station),5)
```

```r
# Your solution code goes here
##Generate a barchart for End Station
top5end_station <- tail(names(sort(table(bikeshare_cities$End.Station))), 5)
common_end_station <- subset(bikeshare_cities, End.Station %in% top5end_station)
common_end_station$End.Station <- factor(common_end_station$End.Station, levels = rev(top5end_station))

ggplot(common_end_station, aes(x = End.Station, stat = 'count', fill = End.Station)) + 
geom_bar() + 
theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Most Common End Stations') +
    xlab('End station') +
    ylab('Count')
head(summary(bikeshare_cities$End.Station),5)
```

```r
##What is the most common trip from start to end (i.e., most frequent combination of start station and end station)?

#Create a Start_to_End Column
Start_to_End <- paste(bikeshare_cities$Start.Station, 'to', bikeshare_cities$End.Station)
```

```r
trips <- data.frame(sort(table(Start_to_End), decreasing = TRUE)[1:5])
names(trips) = c(paste('Most_common_trips', sep=''), 'Count')
trips
```

```r
ggplot(data = trips, aes(stringr::str_wrap(Most_common_trips, 30), Count)) +
    geom_bar(stat = "identity") +  
    theme(axis.text.x = element_text(angle = 90)) +
    ggtitle('Most common trips') +
    xlab('Start station -> End station') +
    ylab('Trip count')    
```

**Summary of your question 3 results goes here.
It can be deduced from the first and second chart that Columbus Circle/ Union Station is both the most common start and end station. While from the third chart, it can be seen that Jefferson Dr & 14th St SW to Jefferson Dr & 14th St SW is the most frequent combination of start and end station.**
